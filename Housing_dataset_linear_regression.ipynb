```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dataset_housing=pd.read_csv("/content/Housing.csv")
dataset_housing
```

```python
model=LinearRegression()
```

```python

dataset_housing['mainroad'] = dataset_housing['mainroad'].replace({'yes': 1, 'no': 0})

dataset_housing['guestroom'] = dataset_housing['guestroom'].replace({'yes': 1, 'no': 0})

dataset_housing['basement'] = dataset_housing['basement'].replace({'yes': 1, 'no': 0})

dataset_housing['hotwaterheating'] = dataset_housing['hotwaterheating'].replace({'yes': 1, 'no': 0})


dataset_housing['airconditioning'] = dataset_housing['airconditioning'].replace({'yes': 1, 'no': 0})


dataset_housing['prefarea'] = dataset_housing['prefarea'].replace({'yes': 1, 'no': 0})
dataset_housing['furnishingstatus'] = dataset_housing['furnishingstatus'].replace({'furnished': 1, 'semi-furnished': -1,'unfurnished':0})
dataset_housing

```

```python
x=dataset_housing[['price','area','bedrooms','bathrooms','stories','mainroad','guestroom','basement','hotwaterheating','airconditioning','parking','prefarea']].values
y=dataset_housing[['furnishingstatus']].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('Coefficient of Determination (R^2):', r2_score(y_test, y_pred))
```

```python
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
```

```python
import numpy as np
import pandas as pd

dataset_housing = pd.read_csv("/content/Housing.csv")


dataset_housing = pd.get_dummies(dataset_housing, columns=['mainroad', 'guestroom', 'basement', 'hotwaterheating', 'airconditioning', 'prefarea', 'furnishingstatus'])

housing_data = dataset_housing.values


def poly_regression_coefficients(dataset, degree):
    X = np.column_stack([dataset[:, 1] ** i for i in range(degree + 1)])
    y = dataset[:, 0]


    coefficients = np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y)

    return coefficients

def predict(coefficients, x):
    y_pred = sum(coefficients[i] * (x ** i) for i in range(len(coefficients)))
    return y_pred

def evaluate_model(test, predictions):
    y_actual = test[:, 0]
    mae = np.mean(np.abs(y_actual - predictions))
    rss = np.sum((y_actual - predictions) ** 2)
    tss = np.sum((y_actual - np.mean(y_actual)) ** 2)
    rsquared = 1 - (rss / tss)
    return {'MAE': mae, 'R-squared': rsquared}


degree = 2

coefficients = poly_regression_coefficients(housing_data, degree)
print('Regression Equation:')
print('y =', end=' ')
for i in range(len(coefficients)):
    print(f'{coefficients[i]}x^{i}', end=' ')
    if i != len(coefficients) - 1:
        print('+', end=' ')
print()

predictions = [predict(coefficients, row[1]) for row in housing_data]
evaluation = evaluate_model(housing_data, predictions)
print('Evaluation Metrics:')
print('MAE:', evaluation['MAE'])
print('R-squared:', evaluation['R-squared'])
```
